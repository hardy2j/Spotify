```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_main=pd.read_csv('/Users/y2z/Desktop/data-2.csv')
df_main=df_main.drop('Unnamed: 0',axis=1)
```

```python
df_acous=df_main[['acousticness','artist','song_title']].sort_values(by='acousticness',ascending=False)[:10]
plt.figure(figsize=(7,7))
sns.barplot(x='song_title',y='acousticness',data=df_acous)
plt.xticks(rotation=90)
plt.show()
```

```python
df_asc_dan=df_main.groupby(['song_title','artist'],as_index=False)[['acousticness','danceability']].sum().sort_values(by=['acousticness','danceability'],ascending=False)[:10]
plt.figure(figsize=(7,7))
sns.barplot(x='song_title',y='danceability',data=df_asc_dan)
plt.xticks(rotation=90)
plt.title('Danceability of the top acoustic songs',size=20)
plt.show()
```

```python
df_asc_dan
```

```python
df_dan_asc=df_main.groupby(['song_title','artist'],as_index=False)[['acousticness','danceability']].sum().sort_values(by=['danceability','acousticness'],ascending=False)[:10]
plt.figure(figsize=(7,7))
sns.barplot(x='song_title',y='acousticness',data=df_dan_asc)
plt.xticks(rotation=90)
plt.title('Acousticness of the top dance worthy songs',size=20)
plt.show()
```

```python
df_dan_loud_t=df_main.groupby(['song_title','artist'],as_index=False)[['danceability','loudness']].sum().sort_values(by=['danceability','loudness'],ascending=False)[:10]
plt.figure(figsize=(7,7))
sns.barplot(x='song_title',y='loudness',data=df_dan_loud_t)
plt.xticks(rotation=90)
plt.title('Loudness of the top dance worthy songs',size=20)
plt.show()
```

```python
df_dan_loud_f=df_main.groupby(['song_title','artist'],as_index=False)[['loudness','danceability']].sum().sort_values(by=['danceability','loudness'],ascending=True)[:10]
plt.figure(figsize=(7,7))
sns.barplot(x='song_title',y='loudness',data=df_dan_loud_f)
plt.xticks(rotation=90)
plt.title('Loudness of the bottom dance worthy songs',size=20)
plt.show()
```
